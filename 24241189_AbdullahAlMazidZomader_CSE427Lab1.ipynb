## NUMPY

```python
import random as rd
import numpy as np
```

```python
# 01
a = np.zeros(20, dtype = int)
b = np.zeros(20, dtype = int)
length = a.size
for i in range(length):
    a[i] = rd.randint(1, 100)
    b[i] = rd.randint(1, 100)
print(f"a: {a}")
print(f"b: {b}")
```

```python
# 02

newA, newB = np.zeros((5, 4), dtype = int), np.zeros((5, 4), dtype = int)

idx = 0
row = newA.shape[0]
col = newA.shape[1]
for i in range(row):
    for ii in range(col):
        newA[i][ii] = a[idx]
        newB[i][ii] = b[idx]
        idx += 1

print(f"New a:\n {newA}")
print(f"New b:\n {newB}")
```

```python
# 03

scalar = newA*newB

print(scalar)
```

```python
# 04

newBT = newB.T

print("Transposed B: \n",newBT)
c = newA @ newBT

print("C: \n",c)


```

```python
# 05

columns_2_3 = c[:, 1:3]
print("2nd and 3rd Column of the matrix C:")
print(columns_2_3)

whole_matrix = c[:, :]
print("\nPrinting whole matrix C using slicing:")
print(whole_matrix)
```

```python
# 06
row, col = c.shape
currMax = c[0][0]
idx = (0, 0)

for i in range(row):
    for ii in range(col):
        if currMax < c[i][ii]:
            currMax = c[i][ii]
            idx = (i, ii)

print(f"Largest Element: {currMax}, Index: [{idx[0]}][{idx[1]}]")
```

```python
# 07

flattenedC = c.flatten()


print("Flatted C: \n", flattenedC)
print("Original C: \n", c)
```

# MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
temperature_dhaka = [25, 34, 21, 45, 28, 6, 43, 18, 7, 2]
humidity_dhaka = [28, 25, 29, 20, 26, 50, 19, 29, 52, 55]

study_hours = [2, 3, 4, 4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks_study = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]

subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks_subjects = [89, 90, 45, 78, 99]
```

```python
# 01
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.xlabel("Temperature (Dhaka)")
plt.ylabel("Humidity (Dhaka)")
plt.title("Temperature vs Humidity")
plt.show()
```

```python
# 02
plt.figure(figsize=(12, 8))
plt.plot(study_hours, marks_study)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Study Hours vs Marks")
plt.show()
```

```python
# 03
plt.barh(subjects, marks_subjects )
plt.xlabel("Marks")
plt.title("Subject-wise Marks (Horizontal Bar)")
plt.show()
```

```python
# 04
plt.bar(subjects, marks_subjects, color=['blue', 'green', 'orange', 'purple', 'brown'])
plt.ylabel("Marks")
plt.title("Subject-wise Marks (Vertical Bar)")
plt.show()
```

```python
# 05
plt.figure(figsize=(16, 12))

# Scatter plot
plt.subplot(2, 2, 1)
plt.scatter(temperature_dhaka, humidity_dhaka, marker='*', color='red')
plt.title("Temperature vs Humidity")

# Line plot (FIXED)
plt.subplot(2, 2, 2)
plt.plot(study_hours, marks_study)
plt.title("Study Hours vs Marks")

# Horizontal bar graph
plt.subplot(2, 2, 3)
plt.barh(subjects, marks_subjects)
plt.title("Horizontal Bar Graph")

# Vertical bar graph
plt.subplot(2, 2, 4)
plt.bar(subjects, marks_subjects, color=['blue', 'green', 'orange', 'purple', 'brown'])
plt.title("Vertical Bar Graph")

plt.tight_layout()
plt.show()
```
